# Density maps

This tutorial shows how to create density maps for 3D point data (works analogously for 2D data). This function uses [vedo](https://vedo.embl.es/) to calculate density maps.

```python
import napari 
import napari_spatial_statistics as nss
from napari.layers import Points, Layer
from napari.utils.io import csv_to_layer_data as cld
from time import perf_counter as pf
```

```python
Oct4 = r"G:\Stardist_results\Point\Oct4_r.csv"
Sox9 = r"G:\Stardist_results\Point\Sox9_r.csv"
```

```python
Oct4p = cld(Oct4)
Sox9p = cld(Sox9)
```

```python
viewer = napari.Viewer(ndisplay=3)
viewer.add_points(Oct4p[0], **Oct4p[1], name='Oct4')
viewer.add_points(Sox9p[0], **Sox9p[1], name='Sox9')
napari.utils.nbscreenshot(viewer, canvas_only=False)
```

Now we can directly calculate a density map from this:

```python
t = pf()

density = nss.density_map(Oct4p[0], radius=100)
viewer.add_image(density, blending='additive', colormap='inferno')
napari.utils.nbscreenshot(viewer, canvas_only=False)

print (f"Perfomance time: {pf() - t:0.2f} seconds.")
```
